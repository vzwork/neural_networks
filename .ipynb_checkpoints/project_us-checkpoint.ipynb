# Problem overview

We want to be able to predict start temperature with neural network based on different obesrvation such as width of the star, the color of the start, etc.

TARGET RMSE = 10000

Numerical features:
- Relational brightness (L/Lo) - relative to the Sun
- Relational radius (R/Ro) - relative to the Sun
- Absolute star mass (Mv)
<br/><br/>

Categorical features:
- Star color - ('white', 'red', 'blue', 'yellow', 'yellow-orange', etc)
- Star type - (0, 1, 2, 3, 4, 5)
  - 0 - Brown dwarf
  - 1 - Red dwarf
  - 2 - White dwarf
  - 3 - Main-sequence star
  - 4 - Super-giant
  - 5 - Hyper-giant
<br/><br/>

Numerical target feature:
- Absolute temperature (T(K)) - temperature on the surface of the star in K (Kelvin)
<br/><br/>

Preliminary analysis:
- Target feature is numerical - we can expect a regression algorithm in the final layer
- Loss function - MSE / MAE
- We will have to test different number of hidden layers
  - Many categorical features (dimensions), so we should try 3-5 hidden layers

```python
def preprocess(df):
    df = df.copy()
    df.color = df.color.replace({
    'Blue-white': 'Blue White',
    'Blue white': 'Blue White',
    'Blue white ': 'Blue White',
    'Blue-White': 'Blue White',
    'yellow-white': 'Yellow',
    'Yellowish White': 'Yellow',
    'White-Yellow': 'Yellow',
    'yellowish': 'Yellow',
    'Yellowish': 'Yellow',
    'Yellow White': 'Yellow',
    'white': 'White',
    'whitish': 'White',
    'Whitish': 'White',
    'Pale yellow orange': 'Orange',
    'Blue ': 'Blue',
    'Orange-Red': 'Orange'
    })
    df.type = df.type.astype('object')
    return df
```

```python
numeric = ['lumin', 'radius', 'mass']
categorical = ['type', 'color']

col_transformer = ColumnTransformer(
                    transformers=[
                        ('scaler', StandardScaler(), numeric),
                        ('ohe', OneHotEncoder(handle_unknown='ignore'), categorical)
                    ], remainder='passthrough')
```

```python
col_transformer.fit(train.drop(columns=['temp']))
```

```python
def process(data, col_transformer):
    return col_transformer.transform(data)
```

```python
valid = preprocess(valid)
test = preprocess(test)
```

```python
X_train = process(train.drop(columns=['temp']), col_transformer)
X_valid = process(valid.drop(columns=['temp']), col_transformer)
X_test = process(test.drop(columns=['temp']), col_transformer)

y_train = train.temp
y_valid = valid.temp
y_test = test.temp
```

# Try Different Neural Networks

Early Stopping Everywhere:
- 1 layer - linear activation function
- 1 layer - ReLU activation function
- 1 layers - HardSwish activation function (SiLU)
- best function 2 layers
- best function 3 layers
- best function 2 layers + dropout
- best function 3 layers + dropout

```python
def train_evaluate(net, X_train, y_train, X_valid, y_valid, patience):
    X_train = torch.FloatTensor(X_train.copy())
    y_train = torch.FloatTensor(y_train.copy().values)
    X_valid = torch.FloatTensor(X_valid.copy())
    y_valid = torch.FloatTensor(y_valid.copy().values)
    
    train_error = []
    valid_error = []
    optimizer = torch.optim.Adam(net.parameters(), lr=0.001)
    optimizer.zero_grad()
    loss = nn.MSELoss()
    
    # early stopping
    best_mse = float('inf')
    counter = patience
    
    while True:
        net.train()
        z_train = net.forward(X_train).flatten()
        mse_train = loss(z_train, y_train)
        train_error.append(mse_train.item())
        
        # learning
        mse_train.backward()
        optimizer.step()
        
        with torch.no_grad():
            net.eval()
            z_valid = net.forward(X_valid).flatten()
            mse_valid = loss(z_valid, y_valid)
            valid_error.append(mse_valid.item())
        
        # early stopping
        if mse_valid < best_mse:
            best_mse = mse_valid
            counter = patience
        else:
            counter -= 1
        if counter == 0:
            break
        
                
    # visualize
    x = np.arange(len(train_error))
    plt.figure(figsize=(4,2))
    plt.plot(x, train_error, color='green')
    plt.plot(x, valid_error, color='red')
    plt.show()
    
    best_mse = min(valid_error)
    best_rmse = best_mse ** 0.5
    print('Best validation MSE: {:.2f}'.format(best_mse))
    print('Best validation RMSE: {:.2f}'.format(best_rmse))
    
    return net, best_rmse
```

 <div style="background: DeepSkyBlue; padding: 2rem;">
    Iterate over different number of neurons.
</div>

```python
print('One linear activation layer')
for n in range(4, 5, 4):
    print('Number of neurons {}'.format(n))
    net = nn.Sequential(*[nn.Linear(X_train.shape[1], n), nn.Linear(n, 1)])
    train_evaluate(net, X_train, y_train, X_valid, y_valid, 4)
    print('---- ---- ---- ---- ---- ---- ---- ---- ---- ')
```

```python
print('Five linear activation layer')
for n in range(4, 21, 4):
    print('Number of neurons {}'.format(n))
    net = nn.Sequential(*[nn.Linear(X_train.shape[1], n),
                          nn.Linear(n, n),
                          nn.Linear(n, n),
                          nn.Linear(n, n),
                          nn.Linear(n, 1)])
    train_evaluate(net, X_train, y_train, X_valid, y_valid, 4)
    print('---- ---- ---- ---- ---- ---- ---- ---- ---- ')
```

```python
print('One ReLU activation layer')
for n in range(8, 25, 8):
    print('Number of neurons {}'.format(n))
    net = nn.Sequential(*[nn.Linear(X_train.shape[1], n), nn.ReLU(), nn.Linear(n, 1)])
    train_evaluate(net, X_train, y_train, X_valid, y_valid, 10)
    print('---- ---- ---- ---- ---- ---- ---- ---- ---- ')
```

```python
print('One SiLU activation layer')
for n in range(32, 45, 4):
    print('Number of neurons {}'.format(n))
    net = nn.Sequential(*[nn.Linear(X_train.shape[1], n), nn.SiLU(), nn.Linear(n, 1)])
    train_evaluate(net, X_train, y_train, X_valid, y_valid, 500)
    print('---- ---- ---- ---- ---- ---- ---- ---- ---- ')
```

```python
print('One Tanh activation layer')
for n in range(24, 41, 8):
    print('Number of neurons {}'.format(n))
    net = nn.Sequential(*[nn.Linear(X_train.shape[1], n), nn.Tanh(), nn.Linear(n, 1)])
    train_evaluate(net, X_train, y_train, X_valid, y_valid, 10)
    print('---- ---- ---- ---- ---- ---- ---- ---- ---- ')
```

```python
print('One Sigmoid activation layer')
for n in range(24, 41, 8):
    print('Number of neurons {}'.format(n))
    net = nn.Sequential(*[nn.Linear(X_train.shape[1], n), nn.Sigmoid(), nn.Linear(n, 1)])
    train_evaluate(net, X_train, y_train, X_valid, y_valid, 10)
    print('---- ---- ---- ---- ---- ---- ---- ---- ---- ')
```

```python
print('Two SiLU activation layers')
for n in range(16, 41, 8):
    print('Number of neurons {}'.format(n))
    net = nn.Sequential(*[nn.Linear(X_train.shape[1], n), 
                          nn.SiLU(), nn.Linear(n, n), 
                          nn.SiLU(), nn.Linear(n, 1)])
    train_evaluate(net, X_train, y_train, X_valid, y_valid, 100)
    print('---- ---- ---- ---- ---- ---- ---- ---- ---- ')
```

```python
print('Two SiLU activation layers + Dropout')
for n in range(24, 41, 8):
    print('Number of neurons {}'.format(n))
    net = nn.Sequential(*[nn.Linear(X_train.shape[1], n), nn.Dropout(0.5), 
                          nn.SiLU(), nn.Linear(n, n),  nn.Dropout(0.5), 
                          nn.SiLU(), nn.Linear(n, 1)])
    train_evaluate(net, X_train, y_train, X_valid, y_valid, 100)
    print('---- ---- ---- ---- ---- ---- ---- ---- ---- ')
```

```python
print('Three SiLU activation layers + Dropout')
for n in range(24, 41, 8):
    print('Number of neurons {}'.format(n))
    net = nn.Sequential(*[nn.Linear(X_train.shape[1], n), nn.Dropout(0.3), 
                          nn.SiLU(), nn.Linear(n, n),  nn.Dropout(0.3), 
                          nn.SiLU(), nn.Linear(n, n),  nn.Dropout(0.3), 
                          nn.SiLU(), nn.Linear(n, 1)])
    train_evaluate(net, X_train, y_train, X_valid, y_valid, 20)
    print('---- ---- ---- ---- ---- ---- ---- ---- ---- ')
```

```python
print('Variable number of neurons and layers')
best_rmse = 10000
best_net = None
for n in range(5, 31, 5):
    print('Number of neurons {}'.format(5))
    setup = []
    setup.append(nn.Linear(X_train.shape[1], 15))
    for i in range(n):
        setup.append(nn.Hardswish())
        setup.append(nn.Linear(15, 15))
    setup.append(nn.Linear(15, 1))
    net = nn.Sequential(*setup)
    net, rmse = train_evaluate(net, X_train, y_train, X_valid, y_valid, 3)
    if rmse < best_rmse:
        best_rmse = rmse
        best_net = net
    print('---- ---- ---- ---- ---- ---- ---- ---- ---- ')
```

```python
best_rmse
```

```python
print('Variable number of neurons and layers')
best_rmse = 10000
best_net = None
for n in range(8, 21, 4):
    print('Number of neurons {}'.format(n))
    setup = []
    setup.append(nn.Linear(X_train.shape[1], n))
    for i in range(int(n/4)):
        setup.append(nn.Hardswish())
        setup.append(nn.Linear(n, n))
    setup.append(nn.Linear(n, 1))
    net = nn.Sequential(*setup)
    net, rmse = train_evaluate(net, X_train, y_train, X_valid, y_valid, 10)
    if rmse < best_rmse:
        best_rmse = rmse
        best_net = net
    print('---- ---- ---- ---- ---- ---- ---- ---- ---- ')
```
